```python
from googletrans import Translator
import pandas as pd

translator = Translator()

file_path = '/home/ttsai/DM_project/Movie-recommendation-predict-main/data_mining_final/Movie-recommendation-predict-1-text-preprocess/data/train_clean.csv'

data = pd.read_csv(file_path)

def translate_row(row):
    ''' Translate elements premise and hypothesis within `row`. '''
    row['new_premise'] = translator.translate(row['premise'], dest='en').text
    row['new_hypothesis'] = translator.translate(row['hypothesis'], dest='en').text
    return row

# 已翻譯的最後一行
last_translated_row = 12150

# 定義每次處理的量跟次數(難以捉模有時可以一次處理1500筆，有時只能150筆QQ)
batch_size = 50
num_iterations = 1

for _ in range(num_iterations):
    # 從已翻譯的最後一行的下一行開始
    start_row = last_translated_row + 1
    end_row = min(last_translated_row + batch_size, len(data))

    # 翻譯到指定行
    data.loc[start_row:end_row] = data.loc[start_row:end_row].apply(translate_row, axis=1)

    # 更新最後的翻譯行
    last_translated_row = end_row

# 更新翻譯結果
data.to_csv(file_path, index=False)
```
